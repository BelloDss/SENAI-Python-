```python
import tkinter

#Gerar um aplicativo - 1 PASSO
app = tkinter.Tk()

# ---------------------------------------------------------------------------------------------------------------------
# funcionalidade do botão -        VARIAVEIS 
def acao():
    v1 = int(input_valor1.get())
    v2 = int(input_valor2.get())
    soma = v1+v2

    txt_resultado.configure(text=f"Resultado:\n{soma}")

# ---------------------------------------------------------------------------------------------------------------------
                                #  NOME, RESOLUÇÃO E TITULO

# Alterar o título do aplicativo - 2 PASSO
app.title ("Calculadora")

# Alterar a resolução inicial - 3 PASSO
#app.geometry("400x470")

# Texto Título - Componente - 4 PASSO
txt_titulo = tkinter.Label(app, text="Somar valores", fg="#f54c00", font=("Helvetica",12))
txt_titulo.grid(column=0, row=0, columnspan=2, padx=60, pady=10)

# ---------------------------------------------------------------------------------------------------------------------

                             # TEXTO ANTES DA CAIXA DE TEXTO/ CAIXA DE TEXTO - PARTE 1

# Texto Valor 1
txt_valor1 = tkinter.Label(app, text="Valor 1:")
txt_valor1.grid(column=0, row=1)

# Entrada Valor 1 (input)
input_valor1 = tkinter.Entry(app) #ENTRY FAZ A CAIXA DE TEXTO 
input_valor1.grid(column=1, row=1)


# ---------------------------------------------------------------------------------------------------------------------

                          # TEXTO ANTES DA CAIXA DE TEXTO/ CAIXA DE TEXTO - PARTE 2

# Texto valor 2
txt_valor2 = tkinter.Label(app, text="Valor 2")
txt_valor2.grid(column=0, row=2)
# input 2
input_valor2 = tkinter.Entry(app)
input_valor2.grid(column=1, row=2, pady=10)

# ---------------------------------------------------------------------------------------------------------------------

# BOTAO 
botao = tkinter.Button (app,text="Somar Valores", command=acao, width=20, font=("Forte",12), fg="White", bg="Blue") #largura do botao - width
botao.grid(column=0, row=3, columnspan=2)

# ---------------------------------------------------------------------------------------------------------------------

# RESULTADO
txt_resultado = tkinter.Label (app, text="")
txt_resultado.grid(column=0,row=4,columnspan=2, pady=20) #pady da um espaçamento- padding

# ---------------------------------------------------------------------------------------------------------------------


#inicar o aplicativo / ULTIMA INSTRUÇÃO 
app.mainloop()
```
